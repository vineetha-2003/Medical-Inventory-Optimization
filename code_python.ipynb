```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#histogram


```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Typeofsales'])
plt.title('Histogram of Numerical Variable')
plt.xlabel('value')
plt.ylabel('frequency')
plt.show()
```

barplot

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df)
plt.title('barplot of categorical variable')
plt.xlabel('category')
plt.ylabel('count')
plt.xticks(rotation=50)
plt.show()
```

scatterplot

```python
plt.figure(figsize=(8, 6))
sns.countplot(df['SubCat1'])
plt.title('count plot of subcat1')
plt.xlabel('Subcat1')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(df['SubCat'])
plt.title('count plot of subcat')
plt.xlabel('Subcat')
plt.ylabel('count')
plt.show()
```

Boxplot

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='ReturnQuantity', y='DrugName', data=df)
plt.title('Box plot of Numerical Variable Across Categories')
plt.xlabel('category')
plt.ylabel('value')
plt.xticks(rotation=50)
plt.show()
```
